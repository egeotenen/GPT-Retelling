### 1. Get Sentiment of Words Without Adverbs

```python
# GET TOKENS ONLY WITHOUT ADVERBS
def tokenize_get_words_only(text):
    tokenized_dict = {}
    tokenized = word_tokenize(text)
    tagged = nltk.pos_tag(tokenized)

    to_remove = set()
    for i in range(len(tagged)):
        if tagged[i][1] in {'RB', 'RBS', 'WRB', 'RBR'}:
            to_remove.add(i)

        new_tagged = [tagged[i] for i in range(len(tagged)) if i not in to_remove]        
        tokens = [tagged[i][0] for i in range(len(tagged)) if i not in to_remove]
    return tokens
```

```python
for column in df.columns:
    df[f'tokenized_{column}_no_adverb'] = df[column].apply(lambda x: tokenize_get_words_only(x))
```

```python
sia = SentimentIntensityAnalyzer()

def sentiment(dataset):
    human_sentiment_compound = []
    for keys in dataset:
        fdist1 = (sia.polarity_scores(keys))
        vocab = list(fdist1.values())[-1] ## [-1] gives compound score
        human_sentiment_compound.append(vocab)
    return statistics.mean(human_sentiment_compound)
```

```python
for column in [col for col in df.columns if col.endswith('no_adverb')]:
    df[f'no_adv_text_{column}'] = df[column].apply(lambda x: ' '.join(x) if isinstance(x, (list, tuple)) else x)
```

```python
for column in [col for col in df.columns if col.endswith('no_adverb')]:
    df[f'sentiment_{column}'] = df[column].apply(lambda x: sentiment(x))
```

### 1.1. Doc2vec - No Adverb
To understand which ones got worse when there is no adverb

```python
model= Doc2Vec.load("test_doc2vec.model")


def sim(mydict):
    doc2vec_sims =[]
    for key in mydict:
        doc2vec_sim = (np.mean(mydict[key]))
        doc2vec_sims.append(doc2vec_sim)
    return doc2vec_sims

def doc(list_sentences):
    mydict={}
    similarities= []
    for i,sentence in enumerate(list_sentences):
        try:
            sentence = sentence.lower()
            words = sentence.split(" ")
            similarities = [model.wv.n_similarity(v1, v2) for v1, v2 in zip(words, words[1:])]
            pairs = [(v1, v2) for v1, v2 in zip(words, words[1:])]

            for a in range(len(similarities)):
                mydict[i]=similarities
        except:
            pass
        
    return sim(mydict)
```

### 4.2. Examine Human Data 

```python
# Get only human data
df_human = df[df['type']=='human']
```

```python
# Calculate pos-tag numbers to get adv
def clean_blank(text):
    clean_text = " ".join(text.split())
    return clean_text

df_human['story'] = df_human['story'].apply(lambda x: clean_blank(x))

### 2. Get Pos-Tags


def create_tokenized(text):
    tokenized = word_tokenize(text)
    tagged = nltk.pos_tag(tokenized)
    return tagged

df_human[f'tokenized'] = df_human['story'].apply(lambda x: create_tokenized(x))

#count the tags
def count_tags(mydata):
    counter_list=[]
    for i in mydata:
        counter_list.append(i[1])

    tag_counts = Counter(counter_list) 
    return tag_counts

df_human['tag_counts'] = df_human['tokenized'].apply(lambda x: count_tags(x))
```

```python
for column in [col for col in df_human.columns if col.endswith('tag_counts')]:
    df_human[f'{column[:-7]}_noun'] = [*(df_human[column].apply(lambda x: (x['NN'] + x['NNS'] + x['NNP']+ x['NNPS'])))]
    df_human[f'{column[:-7]}_verb'] = [*(df_human[column].apply(lambda x: (x['VB'] + x['VBN'] + x['VBG']+ x['VBZ'] + x['VBP']+x['VBD'])))]
    df_human[f'{column[:-7]}_adj'] = [*(df_human[column].apply(lambda x: (x['JJ'] + x['JJS'] + x['JJR'])))]
    df_human[f'{column[:-7]}_adv'] = [*(df_human[column].apply(lambda x: (x['RB'] + x['RBR'] + x['WRB'] +x['RBS'])))]
    df_human[f'{column[:-7]}_pron'] = [*(df_human[column].apply(lambda x: (x['PRP'] + x['PRP$'] + x['WB']+ x['WB$'])))]
    df_human[f'{column[:-7]}_con_det_prep'] = [*(df_human[column].apply(lambda x: (x['DT'] + x['IN'] + x['UH']+ x['TO']+ x['WDT']+ x['EX'])))]
    df_human[f'{column[:-7]}_prep'] = [*(df_human[column].apply(lambda x: (x['IN'])))] 
```
