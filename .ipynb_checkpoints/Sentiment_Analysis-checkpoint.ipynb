### 0. Get Data

```python
df = pd.read_excel('all_gpt_human_only_text.xlsx')

def strip(dataframe):
    return [s.strip() for s in dataframe]

df= df.apply(strip)
```

### 1. Get Sentiment of Words Without Adverbs

```python
# GET TOKENS ONLY WITHOUT ADVERBS
def tokenize_get_words_only(text):
    tokenized_dict = {}
    tokenized = word_tokenize(text)
    tagged = nltk.pos_tag(tokenized)

    to_remove = set()
    for i in range(len(tagged)):
        if tagged[i][1] in {'RB', 'RBS', 'WRB', 'RBR'}:
            to_remove.add(i)

        new_tagged = [tagged[i] for i in range(len(tagged)) if i not in to_remove]        
        tokens = [tagged[i][0] for i in range(len(tagged)) if i not in to_remove]
    return tokens
```

```python
for column in df.columns:
    df[f'tokenized_{column}_no_adverb'] = df[column].apply(lambda x: tokenize_get_words_only(x))
```

```python
sia = SentimentIntensityAnalyzer()

def sentiment(dataset):
    human_sentiment_compound = []
    for keys in dataset:
        fdist1 = (sia.polarity_scores(keys))
        vocab = list(fdist1.values())[-1] ## [-1] gives compound score
        human_sentiment_compound.append(vocab)
    return statistics.mean(human_sentiment_compound)
```

```python
for column in [col for col in df.columns if col.endswith('no_adverb')]:
    df[f'sentiment_{column}'] = df[column].apply(lambda x: sentiment(x))
```

### 2. Get All Words Sentiment 

```python
df_all = pd.read_excel('all_gpt_human_only_text.xlsx')
df_all= df_all.apply(strip)
```

```python
def create_tokenized(text):
    tokenized = word_tokenize(text)
    tagged = nltk.pos_tag(tokenized)
    tokens = [tagged[i][0] for i in range(len(tagged))]
    return tokens
```

```python
for column in df_all.columns:
    df_all[f'tokenized_{column}'] = df_all[column].apply(lambda x: create_tokenized(x))
```

```python
for column in [col for col in df_all.columns if col.startswith('tokenized_')]:
    df_all[f'sentiment_{column}'] = df_all[column].apply(lambda x: sentiment(x))
```

### 3. Prepare Data to Visualize

```python
mean_sentiments = []
for column in [col for col in df_all.columns if col.startswith('sentiment')]:
    mean_sentiments.append(statistics.mean(df_all[column]))
    print(column)
```
